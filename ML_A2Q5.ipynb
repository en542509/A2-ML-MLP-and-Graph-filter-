```python
import numpy as np
import cv2
import scipy.misc
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from keras.utils import np_utils


# For visualization of what is happening inside
from keras.layers import Convolution2D, MaxPooling2D, Flatten

#load image
groupImg = scipy.misc.imread("/home/devin/Desktop/ross_rachel.jpg")/255. 
plt.imshow(groupImg)
plt.show()


```

```python
#kernels from: https://en.wikipedia.org/wiki/Kernel_(image_processing)

# Some kernels from:
# Indentity
kernel_id = [
             [0, 0, 0],
             [0, 1, 0],
             [0, 0, 0]
             ]
kernel_id = np.array(kernel_id,dtype=float)

# Edge detector
kernel_edge = [
             [ 1, 0, -1],
             [ 0, 0,  0],
             [-1, 0,  1]
             ]
kernel_edge = np.array(kernel_edge,dtype=float)
```

```python
group_img_conv_id = np.zeros(groupImg.shape)
for i in range(3):
    group_img_conv_id[:,:,i] = convolve2d(groupImg[:,:,i],kernel_id,
                                        mode='same',
                                        boundary='wrap')
    
#convert to greyscale 
groupImgGray = np.dot(groupImg[...,:3], [0.299, 0.587, 0.114])

# edge detection
groupImgConv = convolve2d(groupImgGray,kernel_edge,
                               mode='same',
                               boundary='wrap')

plt.figure(figsize=(20,10))
plt.subplot(141)
plt.imshow(group_img_conv_id)
plt.title("Original Image")

plt.subplot(142)
plt.imshow(groupImgConv,cmap='gray')
plt.title("Edge detection")


```
